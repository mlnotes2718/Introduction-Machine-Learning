## Simple Linear Regression Formulation (One Feature, b is 0)

### Prepare Environment

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Linear Regression Summary

Let us recall that for linear regression, we need to find the best fit line using the formula below:

$$\hat{y} = b + wx^{(i)}$$

Where:
- $x$ and $y$ are the axis on the graph 
- $b$ is the fixed constant (a.k.a. y-intercept) 
- $w$ is the weight of $x$ (a.k.a gradient of the slope)
	
Or in simpler form

$$f(x^{(i)}) = b + wx^{(i)}$$


### Simple Linear Regression (One Feature, No Intercept)

#### Finding the Best Fit Line

Before we proceed to find the best fit line, we can further simplified the linear regression formula above by assuming $b =0$

If  $b =0$, the the linear regression is as follows:

$$f(x) =wx$$ 

We can use computer to iteratively find many possible fitting lines by computing many possible w. Assuming we have a very small datasets with 3 data points.


## Finding Best Fit Line (Simple 3 Points Example)

When we assume $b = 0$. The formula for linear regression is further simplified as follows:

$$f(x) = wx$$

We also assume a more simple housing dataset with 3 data points, as shown below:

|house size in sq meters|housing price in thousand|
|---|---|
|100|100|
|200|200|
|300|300|

Thus a 100 sq meters (1076 sq ft) cost $100,000

### Ploting 3 Points Data

```python
x = np.array([100,200,300])
y = np.array([100,200,300])
```

```python
# Calculating total number of data points
m = len(x)
print("Number of data points:", m)
```

```python
# Plot data only
fig = plt.figure()
# Plot the data points
plt.scatter(x, y, marker='x', c='r')
# Set the title
plt.title("Simple Dataset")
# Set the y-axis label
plt.ylabel('Selling Price per 1000')
# Set the x-axis label
plt.xlabel('Housing size in sq m')
plt.show()
```

The function to compute our prediction line is as follows:

```python
def compute_prediction_line_simple(x,w):
    m = len(x)
    f = np.zeros(m)
    for i in range(m):
        f[i] = (x[i] * w) 
        
    return f
```

### Ploting Linear Graph if $w=0$

If we predict $w = 0$, then the predicted line show be:

```python
w=0
y_predict1 = compute_prediction_line_simple(x,w)

fig = plt.figure()
# Plot our model prediction
plt.plot(x, y_predict1, c='b',label='Our Prediction w = 0')

# Plot the data points
plt.scatter(x, y, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Simple Dataset")
# Set the y-axis label
plt.ylabel('Selling Price in Thousands')
# Set the x-axis label
plt.xlabel('Housing size in sq m')
plt.legend()
plt.show()
```

### Ploting Linear Graph if $w=5$

If we predict $w = 5$, then the predicted line show be:

```python
w = 5
y_predict2 = compute_prediction_line_simple(x,w)
```

```python
fig = plt.figure()
# Plot our model prediction
plt.plot(x, y_predict2, c='b',label='Our Prediction w = 5')

# Plot the data points
plt.scatter(x, y, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Simple Dataset")
# Set the y-axis label
plt.ylabel('Selling Price in Thousands')
# Set the x-axis label
plt.xlabel('Housing size in sq m')
plt.legend()
plt.show()
```

### Ploting Linear Graph if $w=1$

If we predict $w = 1$, then the predicted line show be:

```python
w = 1
y_predict4 = compute_prediction_line_simple(x,w)

fig = plt.figure()
# Plot our model prediction
plt.plot(x, y_predict4, c='b',label='Our Prediction w = 1')

# Plot the data points
plt.scatter(x, y, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Simple Dataset")
# Set the y-axis label
plt.ylabel('Selling Price in Thousands')
# Set the x-axis label
plt.xlabel('Housing size in sq m')
plt.legend()
plt.show()
```

From above we can know that $w=1$ is the best fit line. 

### Ploting All Linear Graph

The following plot all the best fit line:

```python
fig = plt.figure()
# Plot our model prediction
plt.plot(x, y_predict1, label='Our Prediction w = 0')
plt.plot(x, y_predict2, label='Our Prediction w = 5')
plt.plot(x, y_predict3, label='Our Prediction w = 2')
plt.plot(x, y_predict4, label='Our Prediction w = 1')

# Plot the data points
plt.scatter(x, y, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Simple Dataset")
# Set the y-axis label
plt.ylabel('Selling Price in Thousands')
# Set the x-axis label
plt.xlabel('Housing size in sq m')
plt.legend()
plt.show()
```

Computationally, we can plot many lines from $w = 0$ to $w = 5$

```python
bunch_lines = []

for w in range(0,6):
    print('w is',w)
    line = compute_prediction_line_simple(x,w)
    bunch_lines += [line]

bunch_lines   
```

```python
len(bunch_lines)
```

```python
fig = plt.figure()
# Plot the data points
plt.scatter(x, y, marker='x', c='r',label='Actual Values')

for i in range(len(bunch_lines)):
    plt.plot(x, bunch_lines[i], label=str('Our Prediction w = ' + str(i)))

# Set the title
plt.title("Simple Dataset")
# Set the y-axis label
plt.ylabel('Selling Price in Thousands')
# Set the x-axis label
plt.xlabel('Housing size in sq m')
plt.legend()
plt.show()
```

Plotting all possible lines is only practical for a very small dataset and very simple linear regression. We need to be computationally able to determine w.

To determine best w, we need to know how to compute the difference between the prediction and actual result. We refer to the computation as a cost function. We use the cost function to determine our best fit line while aiming to minimize the cost function.


How to determine the best fit line computationally? We determine the best fit line by calculating the error of our prediction line.
